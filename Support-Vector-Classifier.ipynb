## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## creating datasets
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000, n_features=2,n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(y)
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=.25, random_state=10)
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## n_clusetr_per_class = 2

```python
X,y = make_classification(n_samples=1000, n_features=2,n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=.25, random_state=10)
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf=SVC(kernel='rbf')
rbf.fit(X_train, y_train)
y_pred = rbf.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train, y_train)
y_pred = polynomial.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sigmod=SVC(kernel='sigmoid')
sigmod.fit(X_train, y_train)
y_pred = sigmod.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C':[.1,1,10,100,100],
    'gamma':[1,.1,.01,.001,.0001],
    'kernel':['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred  = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
